```python
import numpy as np
import mdtraj as mdt
import MDAnalysis as mda

from scipy import stats
import matplotlib.pyplot as plt
```

```python
def get_nearby_residues(holo_xtal, ligand_resn, dist_cutoff):

    #handle bricks
    if ligand_resn == "None":
        print("no valid ligands found")
        return []

    #handle everything else
    if type(ligand_resn) == str: #select single ligands
        ligand_select_str = f"resname '{ligand_resn}'"

    elif len(ligand_resn)>1:     #select multiple ligands
        #assemble query for all ligands listed
        molecule_queries = []
        for lname in ligand_resn:
            molecule_queries.append(f"resname '{lname}'")
        ligand_select_str = " or ".join(molecule_queries)

    else:                        #use hardcoded query
        ligand_select_str = ligand_resn[0]

    #get indices of ligand and protein atoms
    ligand_ai = holo_xtal.top.select(f"({ligand_select_str}) and not element H")
    protein_ai = holo_xtal.top.select("protein and not element H")

    #get indices of ligand-coordinating protein atoms; index 0 indicates the 0th trajectory frame
    lining_ai = mdt.compute_neighbors(holo_xtal, dist_cutoff, ligand_ai, haystack_indices=protein_ai, periodic = False)[0]

    #get residue indices and numbers for the holo structure
    lining_resids = np.unique([holo_xtal.top.atom(i).residue.index for i in lining_ai])    #0-indices of the lining residues in the mdtraj structure
    lining_resseqs = np.unique([holo_xtal.top.atom(i).residue.resSeq for i in lining_ai])  #rcsb pdb residue numbers of the lining residues
    
    #print([f"resid {i}" for i in lining_resids])
    lining_ai_resi = np.unique(holo_xtal.top.select(" or ".join([f"resid {i}" for i in lining_resids])))

    lining_ai_byresi = [holo_xtal.top.select(f"resid {i}") for i in lining_resids]
    lining_names_byresi = [str(holo_xtal.top.atom(holo_xtal.top.select(f"resid {i}")[0]).residue) for i in lining_resids]
    
    nonlining_ai_byresi = [holo_xtal.top.select(f"resid {i}") for i in range(holo_xtal.top.n_residues-1) if i not in lining_resids]
    
    return [lining_ai, lining_resids, lining_resseqs, lining_ai_resi, lining_ai_byresi, lining_names_byresi, nonlining_ai_byresi]
```

```python
#filepaths
trjpath = "/Users/jonathanborowsky/Documents/UCSF/classes/bp219/trj"

ptopapo = f"{trjpath}/DESRES-Trajectory_sarscov2-10875753-no-water-zinc-glueCA/sarscov2-10875753-no-water-zinc-glueCA/apo_top.pdb"
ptophol = f"{trjpath}/DESRES-Trajectory_sarscov2-10875754-no-water-zinc-glueCA/sarscov2-10875754-no-water-zinc-glueCA/holo_top.pdb"

ptrjapo  = f"{trjpath}/DESRES-Trajectory_sarscov2-10875753-no-water-zinc-glueCA/sarscov2-10875753-no-water-zinc-glueCA/sarscov2-10875753-no-water-zinc-glueCA-" #0000.dcd
ptrjhol = f"{trjpath}/DESRES-Trajectory_sarscov2-10875754-no-water-zinc-glueCA/sarscov2-10875754-no-water-zinc-glueCA/sarscov2-10875754-no-water-zinc-glueCA-" #0000.dcd
```

```python
#load trajectories
#load all 9 trajectories for each system into a list and then join them

trjsapo = []
trjshol = []

for i in range(0,9):
    trjsapo.append(mdt.load_dcd(ptrjapo + str(i).zfill(4) + ".dcd", top = ptopapo))
    trjshol.append(mdt.load_dcd(ptrjhol + str(i).zfill(4) + ".dcd", top = ptophol))
    
trjapo = mdt.join(trjsapo)
trjhol = mdt.join(trjshol)
```

```python
#get ligand-lining residues using the holo topology
#note that this does not account for changes to 
#ligand coordination over the course of the trajectory

dist_cutoff = 0.5 #nanometers
ligand_resn = "XX5"
tophol = mdt.load(ptophol)

lining_inds = get_nearby_residues(tophol, ligand_resn, dist_cutoff)

#for pymol display:
print("+".join([str(i) for i in lining_inds[2]]))
```

## RMSF calculations and analysis using the entire protein and taking subsets of atoms and residues after calculating RMSF

```python
#calculate RMSF of each atom

rmsf_apo = mdt.rmsf(trjapo, reference = None)
rmsf_hol = mdt.rmsf(trjhol, reference = None)

#separate RMSFs of ligand-lining and non-ligand-lining atoms

rmsf_apo_noligand = [i for x, i in enumerate(rmsf_apo) if x not in lining_inds[3]]
rmsf_apo_ligand = rmsf_apo[lining_inds[3]]
rmsf_hol_noligand = [i for x, i in enumerate(rmsf_hol) if x not in lining_inds[3]]
rmsf_hol_ligand = rmsf_hol[lining_inds[3]]

#calculate rmsfs at the residue level for ligand-lining atoms
rmsf_apo_ligand_byres = []
rmsf_hol_ligand_byres = []

rmsf_apo_noligand_byres = []
rmsf_hol_noligand_byres = []

for r in lining_inds[4]:
    rmsf_apo_ligand_byres.append(np.mean(rmsf_apo[r]))
    rmsf_hol_ligand_byres.append(np.mean(rmsf_hol[r]))

for r in lining_inds[6]:
    rmsf_apo_noligand_byres.append(np.mean(rmsf_apo[r]))
    rmsf_hol_noligand_byres.append(np.mean(rmsf_hol[r]))
```

```python
#plot atom-level rmsf for whole protein

plt.plot(rmsf_apo)
plt.plot(rmsf_hol)
```

```python
#plot atom-level rmsf for ligand-lining residues

plt.plot(rmsf_apo_ligand)
plt.plot(rmsf_hol_ligand)
```

```python
#plot residue-level rmsfs for non-ligand-lining residues
plt.plot(rmsf_apo_noligand_byres)
plt.plot(rmsf_hol_noligand_byres)
```

```python
#plot residue-level rmsfs for ligand-lining residues
plt.plot(rmsf_apo_ligand_byres)
plt.plot(rmsf_hol_ligand_byres)

#label each point with the residue type and number
lining_inds_reformat = [i[0:3]+" "+i[3:] for i in lining_inds[5]]
plt.xticks(range(len(rmsf_apo_ligand_byres)), lining_inds_reformat, size='medium', rotation = 90)

#axis labels
plt.xlabel("residue")
plt.ylabel("RMSF (nm)")

plt.savefig("RMSF-by-resi", dpi = 300, bbox_inches='tight')
```

```python
#t-test to see if the active site residues are significantly more or less flexible
#in the apo vs holo states (2-tailed)

#atom-level
#ligand-lining
twl = stats.ttest_ind(rmsf_apo_ligand, rmsf_hol_ligand, equal_var = False)
#non-ligand-lining
tnl = stats.ttest_ind(rmsf_apo_noligand, rmsf_hol_noligand, equal_var = False)

print(f"ligand-lining residues: {twl}")
print(f"non-ligand-lining residues: {tnl}")

#THE ATOM LEVEL ANALYSIS IS WRONG BECAUSE THE ATOMS DO NOT ALL MOVE INDEPENDENTLY.
#A more sophisticated analysis would take one atom per rotatable bond and use that.
```

```python
#residue-level t-test

#ligand lining
twl = stats.ttest_ind(rmsf_apo_ligand_byres, rmsf_hol_ligand_byres, equal_var = False)

print(f"ligand-lining residues: {twl}")
#the difference is statistically insignificant
```

```python
#residue-level paired t-test

#ligand lining
twlp = stats.ttest_rel(rmsf_apo_ligand_byres, rmsf_hol_ligand_byres)

print(f"ligand-lining residues: {twlp}")
#the difference is statistically insignificant
```
